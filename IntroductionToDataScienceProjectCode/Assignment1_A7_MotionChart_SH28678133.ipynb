# A7.1. Build a Motion Chart comparing the job vacancy rate, the unemployment rate, and the population of each state over time. The motion chart should show the job vacancy rate on the x-axis, the unemployment rate on the y-axis and the bubble size should depend on the population.

```python
from motionchart.motionchart import MotionChart
import pandas as pd
```

```python
# read in the sample data from data.csv
population = pd.read_csv('EstimatedResidentPopulationByStateAndGender.csv')
# have a look at the dataframe, it should be exactly the same as the one we had before
population.head()
```

```python
population.columns = ['Date', '','','','','','','','','','','','','','','','','','','NSW','VIC','QLD','SA','WA','TAS','NT','ACT','']
population.head()
```

```python
# melting population data
population=pd.melt(population,id_vars=['Date'],value_vars=['VIC','QLD','SA','WA','TAS','NT','ACT','NSW'],var_name='States')
population.rename(columns = {'value':'Population'}, inplace = True)
population.head()
```

```python
# getting employmnet data
emp_data=pd.read_excel("EmploymentTimeSeries.xls",sheetname='Time Series')
emp_data.columns = ['State', 'Date', 'employmentRate','unemploymentRate','participiationRate']
emp_data.columns=emp_data.columns.str.strip()
emp_data=emp_data[['State','Date','unemploymentRate']]
emp_data.head()
```

```python
#geting job vacancy data
jobvac_data=pd.read_csv("JobVacancies.csv")
# inverting the date frame
jobvac_data =jobvac_data.transpose()
jobvac_data.head()
```

```python
# changing index of the data
jobvac_data['Date']=jobvac_data.index
jobvac_data.reset_index(drop=True, inplace=True)
jobvac_data.head()
```

```python
# Setting column names, taking them from 1st row
jobvac_data.columns = jobvac_data.iloc[0]

# Removing 1st row
jobvac_data = jobvac_data[1:]
```

```python
jobvac_data.head()
```

```python
# renaming columns
jobvac_data.columns = ['NSW', 'VIC', 'QLD','SA','WA','TAS','NT','ACT','Date']
jobvac_data.head()
```

```python
# now data is in proper format to be melted

# melting job vacancy data
jobvac_data=pd.melt(jobvac_data,id_vars=['Date'],value_vars=['NSW','VIC','QLD','SA','WA','TAS','NT','ACT'],var_name='States')
jobvac_data.rename(columns = {'value':'Vacancies'}, inplace = True)
```

```python
jobvac_data.tail()
```

```python
#merging job vacancy and unemployment data

# changing date format
emp_data.Date=pd.to_datetime(emp_data.Date)
jobvac_data.Date=pd.to_datetime(jobvac_data.Date)

mix_data= pd.merge(emp_data,jobvac_data,left_on=['Date','State'],right_on=['Date','States'])
mix_data.head()
```

```python
# combining population data with other two data

# changing data format of population data
population.Date=pd.to_datetime(population.Date)
population['Date']=population['Date'].dt.strftime('%Y-%d-%m')
population.Date=pd.to_datetime(population.Date)

comb_data=pd.merge(mix_data,population,left_on=['Date','State'],right_on=['Date','States'])
comb_data.head()
```

```python
# creating vacancyRate column
comb_data['VacancyRate']=(comb_data.Vacancies/comb_data.Population)*100
comb_data.head()
```

```python
comb_data.tail()
```

```python
%%html
<style>
.output_wrapper, .output {
    height:auto !important;
    max-height:1000px;  /* your desired max-height here */
}
.output_scroll {
    box-shadow:none !important;
    webkit-box-shadow:none !important;
}
</style>
```

```python
# creating motion chart
motionChart = MotionChart(df = comb_data, key='Date', x='VacancyRate', y='unemploymentRate', xscale='linear', yscale='linear',
                     size='Population', color='State',category='State')

motionChart.to_notebook()
```
