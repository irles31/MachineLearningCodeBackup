# A6.1. Combine the data from the different files into a single table. The table should contain population values, job vacancy counts and unemployment rates for all of the different dates and different States/Territories.


```r
# Getting population data
# Creating a data frame
population_df <- read.csv("EstimatedResidentPopulationByStateAndGender.csv")

head(population_df)

# Changing column names ---------

names(population_df)[names(population_df) == 'X'] <- 'Date'

names(population_df)[names(population_df) == 'Estimated.Resident.Population....Persons....New.South.Wales..'] <- 'NSW'
names(population_df)[names(population_df) == 'Estimated.Resident.Population....Persons....Victoria..'] <- 'VIC'
names(population_df)[names(population_df) == 'Estimated.Resident.Population....Persons....Queensland..'] <- 'QUE'
names(population_df)[names(population_df) == 'Estimated.Resident.Population....Persons....South.Australia..'] <- 'SA'
names(population_df)[names(population_df) == 'Estimated.Resident.Population....Persons....Western.Australia..'] <- 'WA'
names(population_df)[names(population_df) == 'Estimated.Resident.Population....Persons....Tasmania..'] <- 'TAS'
names(population_df)[names(population_df) == 'Estimated.Resident.Population....Persons....Northern.Territory..'] <- 'NT'
names(population_df)[names(population_df) == 'Estimated.Resident.Population....Persons....Australian.Capital.Territory..'] <- 'ACT'

population_data <- subset(population_df, select = c(Date, NSW, VIC, QUE, SA, WA, TAS, NT, ACT))
#converting Date from factor to date
population_data$Date <- as.Date(population_df$Date, format = "%d/%m/%Y")
head(population_data)
tail(population_data)

#install.packages("reshape")
library(reshape)
pop_data <- melt(population_data, id=c("Date"))
names(pop_data)[names(pop_data) == 'variable'] <- 'State'
names(pop_data)[names(pop_data) == 'value'] <- 'population'
head(pop_data)
```

```r
# Extracting job vacany data --------------
job_vac <- read.csv("JobVacancies.csv")
head(job_vac)
jobvac_inv <- setNames(data.frame(t(job_vac[,-1])), job_vac[,1])
head(jobvac_inv)
#jobvac_inv[,1]
library(data.table)
setDT(jobvac_inv, keep.rownames = TRUE)[]
#list(jobvac_inv)
names(jobvac_inv)[names(jobvac_inv) == 'rn'] <- 'Date'
#cleaning date column
jobvac_inv$Date <- substring(jobvac_inv$Date, 2)
jobvac_inv$Date <- as.Date(jobvac_inv$Date, format = "%Y.%m.%d")
head(jobvac_inv)
vac_data <- melt(jobvac_inv, id=c("Date"))
names(vac_data)[names(vac_data) == 'variable'] <- 'State'
names(vac_data)[names(vac_data) == 'value'] <- 'jobVacancy'
head(vac_data)
```

```r
# Extracting unemployment rate data ---------------
library(readxl)
# list the sheets of the excel file
excel_sheets("EmploymentTimeSeries.xls")
# Call info from the sheets using read_excel
empTimeSeries <- read_excel('EmploymentTimeSeries.xls',sheet='Time Series')
names(empTimeSeries) <- c("State", "Date", "EmploymentRate", "UnemploymentRate", "ParticipationRate")
head(empTimeSeries)
#unemployment data ------------------------------
unedf <- subset(empTimeSeries, select = -c(EmploymentRate, ParticipationRate))
head(unedf)
```

```r
#merging 3 data sets on basis of date

comb_data <- merge(pop_data, vac_data, by=c("Date", "State"))
#head(comb_data)
combined_data <- merge(comb_data, unedf, by=c("Date", "State")) 
head(combined_data)
tail(combined_data)
```
